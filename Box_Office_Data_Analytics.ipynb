## Import Libraries

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
movies=pd.read_csv("/content/box_office.csv")
```

## Task 2: Visualizing the Target Distribution

```python

most_profitable_genre = pd.DataFrame(movies,columns=['all_genres','revenue'])
plt.figure(figsize=(100,25))
ax = sns.barplot(x=most_profitable_genre['all_genres'],y=most_profitable_genre['revenue'])
x=ax.set_xlabel("Movie Genre")
b=ax.set_ylabel("Revenue (in Million Dollars)")
c=ax.set_xticklabels(labels=ax.get_xticklabels(),rotation=90)
d=ax.set_title("Most Profitable Movies by Genre")
```

```python
most_popular_genre = pd.DataFrame(movies,columns=['all_genres','popularity'])
plt.figure(figsize=(100,25))
ax = sns.barplot(x=most_popular_genre['all_genres'],y=most_popular_genre['popularity'])
x=ax.set_xlabel("Movie Genre")
b=ax.set_ylabel("Average Popularity Points")
c=ax.set_xticklabels(labels=ax.get_xticklabels(),rotation=90)
d=ax.set_title("Most Popular Movies by Genre")
```

## Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(movies['budget'], movies['revenue'])
plt.title('Revenue vs budget');
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()
```

## Task 4: Distribution of Languages in Film

```python
language_data = movies.loc[movies['original_language'].isin(movies['original_language'].value_counts().head(10).index)]
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 1)
sns.boxplot(x='original_language', y = 'revenue', data=language_data)
plt.title('Revenue per language')
```
